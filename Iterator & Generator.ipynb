```
# it is both iterable and iterator
class yrange:
  # n is the number to which i want a range
  def __init__(self, n):
    self.i = 0
    self.n = n
    # method makes our class iterable
  def __iter__(self):
    return self
  def __next__(self):
    if self.i < self.n:
      i=self.i
      self.i+=1
      return i
    else:
      raise StopIteration()
```

```
c = yrange(7)
```

```
list(c)
```

```
list(c)
```

```
for x in yrange(2):
  print(x)
```

```
# Iterable class
class zrange:
  def __init__(self,n):
    self.n=n
  def __iter__(self):
    return zrange_iter(self.n)
# Iterator class
class zrange_iter:
  def __init__(self,n):
    self.i = 0
    self.n = n
  def __iter__(self):
    return self
  def __next__(self):
    if self.i < self.n:
      i=self.i
      self.i+=1
      return i
    else:
      raise StopIteration()
```

```
def fib():
  prev,cur=0,1
  while True:
    yield cur
    prev,cur = cur, prev + cur
```

```
type(fib())
```

```
g = fib()
```

```
next(g)
```
