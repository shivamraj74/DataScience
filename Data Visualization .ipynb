```
import matplotlib.pyplot as plt
```

```
import numpy as np
```

```
x = np.arange(10)
```

```
y1=x**2
y2=2*x+3
```

```
plt.style.use('seaborn')
```

```
plt.plot(x,y1)
plt.show()
```

```
# Line Plots
plt.plot(x,y1,color='purple',label="sc",marker='o')
plt.plot(x,y2,color='red',label="sr",linestyle="dashed",marker='*')
plt.xlabel("Time")
plt.ylabel("Knowledge")
plt.title("Knowledge of Person over Time")
plt.legend()
plt.show()
```

```
# Scatter Plots 

plt.figure(figsize=(6,6))
plt.scatter(x,y1,color='red',label="sr",marker='^')
plt.scatter(x,y2,color='green',label="sc",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Knowledge")
plt.title("Scatter Plots - Knowledge Over Time")
plt.legend()
plt.show()
```

```
plt.style.use("seaborn")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"]) # Current Year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year") # Next Year
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

```
plt.style.use("dark_background")

plt.figure(figsize=(5,5))
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"],color="yellow") # Current Year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year",color="orange") # Next Year
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.legend()
plt.show()
```

```
plt.figure(figsize=(3,3))
subjects = "DSA","DBMS","OS","CN"
weightage = [20,10,15,5]

plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),autopct='%1.1f%%',shadow=True)
plt.title("Subjects ")
plt.show()
```

```
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)
X2 = np.round(Xsn*7 + 40)

print(X,X2)
```

```
plt.style.use("seaborn")

plt.hist(X2,alpha=0.8,label="DSA")
plt.hist(X,alpha=0.9,label="DBMS")
plt.ylabel("Prob/FreqCount of Students")
plt.xlabel("Marks Range")
plt.title("Histogram")
plt.legend()
plt.show()
```
