```
# Gradient Descent
```

```
import numpy as np
import matplotlib.pyplot as plt

```

```
X = np.arange(10)
Y = (X-5)**2
```

```
print(X,Y)
```

```
# Given a f(x), find x that minimizes f
plt.style.use("seaborn")
plt.plot(X,Y)
plt.show()
```

```
plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

```
x = 0
lr = 0.1
error = []
# 50 steps in the downhill direction
plt.plot(X,Y)

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```
# Plot the values of error

plt.plot(error)
plt.show()
```
